```python
import numpy as np

def sigmoid(x):
    """
    Compute the sigmoid function for the given input.
    
    Arguments:
    x -- A scalar or numpy array of any shape.
    
    Returns:
    s -- Output of the sigmoid function, same shape as input x.
    """
    s = 1 / (1 + np.exp(-x))
    return s
```

```python
# SVM - Regression - linear 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the SVM regression model
def train_svm_regression(X_train, y_train):
    model = SVR(kernel='linear')  # You can try different kernels, e.g., 'rbf', 'poly', etc.
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("last_3.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the SVM regression model
    model = train_svm_regression(X_train, y_train)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)

    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
import random

def generate_random_floats_bw_nos(length, start, end):
    if not isinstance(length, int) or length <= 0:
        raise ValueError("Length must be a positive integer.")

    if not (isinstance(start, float) and isinstance(end, float)):
        raise ValueError("Start and end values must be floating-point numbers.")

    if start >= end:
        raise ValueError("Start value must be less than the end value.")

    random_floats = [random.uniform(start, end) for _ in range(length)]
    return random_floats

if __name__ == "__main__":
    num_random_floats = 5
    start_value = 0.0
    end_value = 1.0

    random_floats_list = generate_random_floats_bw_nos(num_random_floats, start_value, end_value)
    print("Random floating-point numbers between {} and {}:".format(start_value, end_value))
    print(random_floats_list)
```

```python
# Neural Networks
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the Neural Network model
def create_and_train_neural_network(X_train, y_train):
    model = Sequential()
    model.add(Dense(64, input_dim=X_train.shape[1], activation='relu'))
    model.add(Dense(32, activation='relu'))
    model.add(Dense(1, activation='linear'))

    model.compile(loss='mean_squared_error', optimizer='adam')
    model.fit(X_train, y_train, epochs=50, batch_size=32, verbose=1)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("last_12_10k.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the Neural Network model
    model = create_and_train_neural_network(X_train, y_train)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Calculate evaluation metrics for regression
    mse = np.mean((y_test - y_pred) ** 2)
    rmse = np.sqrt(mse)
    r2 = 1 - (np.sum((y_test - y_pred) ** 2) / np.sum((y_test - np.mean(y_test)) ** 2))

    print("Mean Squared Error:", mse)
    print("Root Mean Squared Error:", rmse)
    print("R-squared:", r2)
```

```python
# K Means Clustering 
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def kmeans_clustering(data, num_clusters=3):
    # Initialize the KMeans model
    kmeans = KMeans(n_clusters=num_clusters, random_state=42)

    # Fit the model to the data
    kmeans.fit(data)

    # Get the cluster labels for each data point
    cluster_labels = kmeans.labels_

    # Calculate the silhouette score
    silhouette_avg = silhouette_score(data, cluster_labels)

    # Create a list of lists to store indices of data points in each cluster
    clusters_indices = [[] for _ in range(num_clusters)]

    # Fill the cluster indices list
    for idx, label in enumerate(cluster_labels):
        clusters_indices[label].append(idx)

    return clusters_indices, silhouette_avg

if __name__ == "__main__":
    # Read data from the CSV file into a DataFrame
    data_file = "last_15.csv"  # Replace with the actual path to your CSV file
    data = pd.read_csv(data_file)

    # Specify the columns containing the features you want to use for clustering
    feature_columns = ["load_ind", "speed_ind","recomm_tyre_press"]  # Replace with your actual feature column names

    # Extract the feature data from the DataFrame
    data_points = data[feature_columns].values

    # Number of clusters
    num_clusters = 3

    # Call the k-means clustering function
    clusters_indices, silhouette_score_value = kmeans_clustering(data_points, num_clusters)

    # Print the results
    print("Cluster indices:")
    print(clusters_indices)

    print("Silhouette score:", silhouette_score_value)
```

```python
#Gradient Boosting Machine - Regression 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score


def read_data(filename):
    data = pd.read_csv(filename)
    return data


def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y


def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test


def train_gradient_boosting_regression(X_train, y_train):
    model = GradientBoostingRegressor()
    model.fit(X_train, y_train)
    return model


def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred


def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    
    data = read_data("last_30.csv")

    
    X, y = split_data(data)

    
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    
    model = train_gradient_boosting_regression(X_train, y_train)

    
    y_pred = make_predictions(model, X_test)

    
    mse, r2 = evaluate_model(y_test, y_pred)
    print("Gradient Boosting Machine - Regression (last_30.csv)")
    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```
