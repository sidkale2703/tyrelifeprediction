```python
# MULTIVARIATE LINEAR REGRESSION
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression 
from sklearn.metrics import mean_squared_error,r2_score

def read_data(filename):
    data = pd.read_csv(filename)
    return data

def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:,-1]
    return X,y

def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

def train_linear_regression(X_train,y_train):
    model = LinearRegression()
    model.fit(X_train,y_train)
    return model

def make_predictions(model,X_test):
    y_pred = model.predict(X_test)
    return y_pred
def evaluate_model(y_test,y_pred):
    mse = mean_squared_error(y_test,y_pred)
    r2 = r2_score(y_test,y_pred)
    return mse,r2

if __name__=="__main__":
    data = read_data("recomm_tyre_press.csv")
    X,y = split_data(data)
    X_train,X_test,y_train,y_test = split_train_test(X,y)
    model = train_linear_regression(X_train,y_train)
    y_pred = make_predictions(model,X_test)
    mse,r2 = evaluate_model(y_test,y_pred)
    print("Mean Squared Error:",mse)
    print("R-squared:",r2)
```

```python
# K NEAREST NEIGHBOURS - REGRESSION
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the KNN Regression model
def train_knn_regression(X_train, y_train, n_neighbors=5):
    model = KNeighborsRegressor(n_neighbors=n_neighbors)
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("recomm_tyre_press.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the KNN Regression model
    model = train_knn_regression(X_train, y_train, n_neighbors=5)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)

    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
# Gradient Boosting Machine - Regression 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the Gradient Boosting Machine Regression model
def train_gradient_boosting_regression(X_train, y_train):
    model = GradientBoostingRegressor()
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("recomm_tyre_press.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the Gradient Boosting Machine Regression model
    model = train_gradient_boosting_regression(X_train, y_train)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)

    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
# GRADIENT BOOSTING MACHINE - REGRESSION 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the Gradient Boosting Machine Regression model
def train_gradient_boosting_regression(X_train, y_train):
    model = GradientBoostingRegressor()
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("final30.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the Gradient Boosting Machine Regression model
    model = train_gradient_boosting_regression(X_train, y_train)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)
    print("Gradient Boosting Machine - Regression")
    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
# KNN - REGRESSION 
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the KNN Regression model
def train_knn_regression(X_train, y_train, n_neighbors=5):
    model = KNeighborsRegressor(n_neighbors=n_neighbors)
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("final30.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the KNN Regression model
    model = train_knn_regression(X_train, y_train, n_neighbors=5)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)
    print("KNN - Regression")
    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
# K means clustering
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Load the data from the CSV file
data = pd.read_csv("final30.csv")

# Step 2: Prepare the data for clustering
X = data.values  # Convert DataFrame to a NumPy array for clustering

# Step 3: Create and fit the KMeans model with 2 clusters
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Step 4: Get the cluster assignments and cluster centers
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Step 5: Print the indexes belonging to each cluster and their percentage of belongingness to cluster 1
for i, cluster_label in enumerate(cluster_labels):
    cluster_center_1 = cluster_centers[0]  # Center of cluster 1
    cluster_center_2 = cluster_centers[1]  # Center of cluster 2
    distance_to_cluster_1 = np.linalg.norm(X[i] - cluster_center_1)  # Distance to cluster 1 center
    distance_to_cluster_2 = np.linalg.norm(X[i] - cluster_center_2)  # Distance to cluster 2 center
    percentage_cluster_1 = distance_to_cluster_1 / (distance_to_cluster_1 + distance_to_cluster_2) * 100
    print(f"Index {i}: Belongs {percentage_cluster_1:.2f}% to cluster 1")

# Step 6: Calculate and print the silhouette score for 2 clusters
silhouette_avg = silhouette_score(X, cluster_labels)
print(f"\nSilhouette Score for 2 clusters: {silhouette_avg:.4f}")
```
