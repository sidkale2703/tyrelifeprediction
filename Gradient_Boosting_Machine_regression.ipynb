```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import learning_curve
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score


data = pd.read_csv("rem_life_km_10000.csv")  

X = data.iloc[:, :-1]  
y = data.iloc[:, -1]    


n_estimators = 75
learning_rate = 0.025
max_depth = 3
min_samples_split = 2
min_samples_leaf = 1
subsample = 1.0
loss = 'ls'  


model = GradientBoostingRegressor(
    n_estimators=n_estimators,
    learning_rate=learning_rate,
    max_depth=max_depth,
    min_samples_split=min_samples_split,
    min_samples_leaf=min_samples_leaf,
    subsample=subsample,
    loss=loss,
    random_state=42
)


train_sizes, train_scores, cv_scores = learning_curve(
    model, X, y, cv=5, train_sizes=np.linspace(0.1, 1.0, 10), n_jobs=-1
)


train_scores_mean = np.mean(train_scores, axis=1)
cv_scores_mean = np.mean(cv_scores, axis=1)
train_scores_std = np.std(train_scores, axis=1)
cv_scores_std = np.std(cv_scores, axis=1)


for i, train_size in enumerate(train_sizes):
    print(f"Train Size: {train_size}")
    print(f"  Training Score: {train_scores_mean[i]} (+/- {train_scores_std[i]})")
    print(f"  Cross-validation Score: {cv_scores_mean[i]} (+/- {cv_scores_std[i]})")
    print("train - cross_val  = ", train_scores_mean[i] - cv_scores_mean[i])


model.fit(X, y)
y_pred = model.predict(X)


mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)


print("\n")
print(f"Mean Squared Error:", mse)
print(f"R-squared Score:", r2)


plt.figure(figsize=(10,6))
plt.title('Learning Curve')
plt.xlabel('Number of Training Examples')
plt.ylabel('Score')
plt.grid()

plt.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color='r')
plt.fill_between(train_sizes, cv_scores_mean - cv_scores_std, cv_scores_mean + cv_scores_std, alpha=0.1, color='g')

plt.plot(train_sizes, train_scores_mean, 'o-', color='r', label='Training score')
plt.plot(train_sizes, cv_scores_mean, 'o-', color='g', label='Cross-validation score')

plt.legend(loc='best')
plt.show()
```
