```python
# K means clustering
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Step 1: Load the data from the CSV file
data = pd.read_csv("mid_1.csv")

# Step 2: Prepare the data for clustering
X = data.values  # Convert DataFrame to a NumPy array for clustering

# Step 3: Create and fit the KMeans model with 2 clusters
num_clusters = 2
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(X)

# Step 4: Get the cluster assignments and cluster centers
cluster_labels = kmeans.labels_
cluster_centers = kmeans.cluster_centers_

# Step 5: Print the indexes belonging to each cluster and their percentage of belongingness to cluster 1
for i, cluster_label in enumerate(cluster_labels):
    cluster_center_1 = cluster_centers[0]  # Center of cluster 1
    cluster_center_2 = cluster_centers[1]  # Center of cluster 2
    distance_to_cluster_1 = np.linalg.norm(X[i] - cluster_center_1)  # Distance to cluster 1 center
    distance_to_cluster_2 = np.linalg.norm(X[i] - cluster_center_2)  # Distance to cluster 2 center
    percentage_cluster_1 = distance_to_cluster_1 / (distance_to_cluster_1 + distance_to_cluster_2) * 100
    print(f"Index {i}: Belongs {percentage_cluster_1:.2f}% to cluster 1")

# Step 6: Calculate and print the silhouette score for 2 clusters
silhouette_avg = silhouette_score(X, cluster_labels)
print(f"\nSilhouette Score for 2 clusters: {silhouette_avg:.4f}")
```

```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

def read_csv_data(file_path):
    return pd.read_csv(file_path)

def k_means_clustering(data, num_clusters=3):
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(data)
    cluster_labels = kmeans.labels_
    centroids = kmeans.cluster_centers_
    return cluster_labels, centroids

def calculate_belongs_to_each_cluster(data, cluster_labels, num_clusters=3):
    belongs_to_clusters = np.zeros((len(data), num_clusters))
    for i, label in enumerate(cluster_labels):
        belongs_to_clusters[i, label] = 1
    return belongs_to_clusters

def print_belongs_to_each_cluster(belongs_to_clusters):
    print("Percentage of Belongingness to Each Cluster:")
    for idx, belongs_to_cluster in enumerate(belongs_to_clusters):
        percentages = belongs_to_cluster * 100
        print(f"Index {idx}: Cluster 0: {percentages[0]:.2f}%, Cluster 1: {percentages[1]:.2f}%, Cluster 2: {percentages[2]:.2f}%")

def calculate_silhouette_score(data, cluster_labels):
    return silhouette_score(data, cluster_labels)

if __name__ == "__main__":
    # Replace 'data.csv' with the actual file path of your CSV file
    file_path = "mid_2.csv"

    # Read data from CSV
    data = read_csv_data(file_path)

    # Perform K-means clustering
    cluster_labels, centroids = k_means_clustering(data)

    # Calculate percentage of belongingness to each cluster
    belongs_to_clusters = calculate_belongs_to_each_cluster(data, cluster_labels)

    # Print percentage of belongingness to each cluster for each index
    print_belongs_to_each_cluster(belongs_to_clusters)

    # Calculate and print the silhouette score
    silhouette_score_val = calculate_silhouette_score(data, cluster_labels)
    print(f"Silhouette Score: {silhouette_score_val:.4f}")
```

```python
# GRADIENT BOOSTING MACHINE - REGRESSION
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the Gradient Boosting Machine Regression model
def train_gradient_boosting_regression(X_train, y_train):
    model = GradientBoostingRegressor()
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("percent_hard.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the Gradient Boosting Machine Regression model
    model = train_gradient_boosting_regression(X_train, y_train)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)

    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Read the data from the CSV file
def read_data(filename):
    data = pd.read_csv(filename)
    return data

# Step 2: Split the data into features (X) and target (y)
def split_data(data):
    X = data.iloc[:, :-1]
    y = data.iloc[:, -1]
    return X, y

# Step 3: Split the data into training and testing sets
def split_train_test(X, y, test_size=0.2, random_state=42):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=random_state)
    return X_train, X_test, y_train, y_test

# Step 4: Create and train the Random Forest Regression model
def train_random_forest_regression(X_train, y_train):
    model = RandomForestRegressor()
    model.fit(X_train, y_train)
    return model

# Step 5: Make predictions using the trained model
def make_predictions(model, X_test):
    y_pred = model.predict(X_test)
    return y_pred

# Step 6: Evaluate the model's performance
def evaluate_model(y_test, y_pred):
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    return mse, r2

if __name__ == "__main__":
    # Step 1: Read the data from the CSV file
    data = read_data("new_1.csv")

    # Step 2: Split the data into features (X) and target (y)
    X, y = split_data(data)

    # Step 3: Split the data into training and testing sets
    X_train, X_test, y_train, y_test = split_train_test(X, y)

    # Step 4: Create and train the Random Forest Regression model
    model = train_random_forest_regression(X_train, y_train)

    # Step 5: Make predictions using the trained model
    y_pred = make_predictions(model, X_test)

    # Step 6: Evaluate the model's performance
    mse, r2 = evaluate_model(y_test, y_pred)

    print("Mean Squared Error:", mse)
    print("R-squared:", r2)
```

```python
import pandas as pd
import pickle

# Load the trained model from the file (change 'model_file.pkl' to the actual file path)
model_filename = 'model_file_total_life_km.pkl'
with open(model_filename, 'rb') as file:
    model = pickle.load(file)

# Load the input data from the separate CSV file
input_data = pd.read_csv("final33.csv")

# Assuming the two input variables are in columns 'variable1' and 'variable2'
# Extract the input variables from the input data
variable1_values = input_data['speed_rating_kmph']
variable2_values = input_data['load_ind_kg']
variable3_values = input_data['vehicle_speed_kmph']
variable4_values = input_data['vehicle_weight_kg']
variable5_values = input_data['tyre_pressure_psi']
variable6_values = input_data['recomm_tyre_press_psi']
variable7_values = input_data['tread_depth_mm_new']



# Create a list of input pairs for prediction
input_vars = list(zip(variable1_values, variable2_values, variable3_values, variable4_values,variable5_values,variable6_values,variable7_values))

# Make predictions
predicted_output = model.predict(input_vars)

# Print the predicted output
print("Predicted target variables:", predicted_output)
```

```python
def round_to_nearest_hundred(number):
    return round(number, -2)
```

```python
prd_out_total_life_km = []
for i in predicted_output:
    prd_out_total_life_km.append(round_to_nearest_hundred(i))
print(len(prd_out_total_life_km))
print(prd_out_total_life_km)
```
